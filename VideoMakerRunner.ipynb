```python
# Test split functionality

scenarioOptions = {
    "results" : "Results.Test",
    "shape" : (720, 1280, 3), 
    "current_index" : 20, 
    "step" : 10, 
    "n_frames" : 30
}

vm = VideoMaker(scenarioOptions)

# 0.5 sec text, 0.5 image
sc0 = [{
    "type" : "text",
    "description" : "Text 0...",
    "duration" : 0.5,
    "text": ["Text 0"]
}, {
    "type" : "image",
    "description" : "catsm.jpeg",
    "duration" : 0.5,
    "image": "Images/catsm.jpeg"
}]

# 1 sec text
sc1 = [{
    "type" : "text",
    "description" : "Text 1...",
    "duration" : 1,
    "text": ["Text 1"]
}]

# 0.5 sec image, 0.5 text
sc2 = [{
    "type" : "image",
    "description" : "catsm.jpeg",
    "duration" : 0.5,
    "image": "Images/catsm.jpeg"
},
    {
    "type" : "text",
    "description" : "Text 2...",
    "duration" : 0.5,
    "text": ["Text 2"]
}]

# a generator of text
sc3 = vm.scn_text_slow_gen(["Hello", "World!"], 
                        scene_options={"mode" : "center", "bgcolor":(0, 0, 128), "color":(255, 255, 0)})

split = {
    "shape" : [[0, 1], [2, 3]],
    "generators": {
        0: sc0,
        1: sc1,
        2: sc2,
        3: sc3
    },
    "mode" : "min"
}

for (cix, i), img in vm.scn_split_gen(scene_options=split):
    mpimg.imsave(vm.results + "/frame%04d.%04d.jpeg" % (cix, i), img)

fig = plt.figure()
plt.imshow(img)
```

```python
# Test cut functionality

scenarioOptions = {
    "results" : "Results.Test",
    "shape" : (720, 1280, 3), 
    "current_index" : 20, 
    "step" : 10, 
    "n_frames" : 30
}

vm = VideoMaker(scenarioOptions)

# background process - a scenario of an image
bg = [{
    "type" : "image",
    "description" : "bg.jpeg",
    "duration" : 2,
    "image": "Images/map1799.jpeg"
}]

# half text, half image
sc1 = [{
    "type" : "text",
    "description" : "Text 0...",
    "duration" : 1,
    "text": ["Text 0"]
}, {
    "type" : "image",
    "description" : "catsm.jpeg",
    "duration" : 1,
    "image": "Images/CatSm.jpeg"
}]

# a generator of text
sc2 = vm.scn_text_slow_gen(["Hello", "World!"], 
                        scene_options={"mode" : "center", "bgcolor":(0, 0, 128), "color":(255, 255, 0)})

# definition of cut: screen is split in 4 parts. sc1 goes inot top right corner. sc2 goes into bottom left corner.
cut = {
    "shape" : [[0, 1], [2, 3]],
    "bg" : bg,
    "generators": {
        1: sc1,
        2: sc2
    }
}

for (cix, i), img in vm.scn_cut_gen(scene_options=cut):
    mpimg.imsave(vm.results + "/frame%04d.%04d.jpeg" % (cix, i), img)

fig = plt.figure()
plt.imshow(img)
```

```python
# test zoom functionality

scenarioOptions = {
    "results" : "Results.Test",
    "shape" : (720, 1280, 3), 
    "current_index" : 20, 
    "step" : 10, 
    "n_frames" : 30
}

vm = VideoMaker(scenarioOptions)

bg = vm.scn_text_slow_gen(["Lorem ipsum dolor sit amet", 
                           "Consectetur adipiscing elit", 
                           "Sed do eiusmod tempor incididunt"], 
                        scene_options={"mode" : "center", "bgcolor":(0, 0, 128), "color":(255, 255, 0)})

sc = vm.scn_text_slow_gen(["The quick brown fox jumps over the lazy dog. Or does she?"], 
                        scene_options={"mode" : "center", "bgcolor":(0, 128, 128), "color":(255, 255, 0)})

cut_tl = {
    "duration" : 1,
    "shape" : [[0, 1], [2, 3]],
    "bg" : bg,
    "generators": {
        0: sc
    }
}

cut_tr = {
    "duration" : 1,
    "shape" : [[0, 1], [2, 3]],
    "bg" : bg,
    "generators": {
        1: sc
    }
}

zoom_in = {
    "bg" : bg,
    "fg": sc,
    "delay" : 20
}

zoom_out_tl = {
    "bg" : bg,
    "fg": sc,
    "in_out" : "out",
    "delay" : 20
}

zoom_out_tr = {
    "bg" : bg,
    "fg": sc,
    "in_out" : "out",
    "delay" : 20,
    "corner" : "tr"
}

# A bunch of cuts and zooms: cut in tl corner, zoom-in, zoom out in the same corner, zoom in, zoom out in tr corner.

vm.append({
    "type" : "cut",
    "description" : "Cut",
    "options" : cut_tl
})

vm.append({
    "type" : "zoom",
    "description" : "Zoom In",
    "options": zoom_in
})

vm.append({
    "type" : "zoom",
    "description" : "Zoom Out",
    "options": zoom_out_tl
})

vm.append({
    "type" : "cut",
    "description" : "Cut",
    "options" : cut_tl
})

vm.append({
    "type" : "zoom",
    "description" : "Zoom In",
    "options": zoom_in
})

vm.append({
    "type" : "zoom",
    "description" : "Zoom Out",
    "options": zoom_out_tr
})

vm.append({
    "type" : "cut",
    "description" : "Cut",
    "options" : cut_tr
})

vm.process()
```
